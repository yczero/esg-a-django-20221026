```python
from diary.models import Memory
```

```python
# 데이터베이스에 즉시 저장
Memory.objects.create(title="title#1", message = "content#1")
```

```python
memory = Memory()
memory.title = "title#2"
memory.message = "content#2"
memory.save()   #save 호출시점에 저장
```

```python
memory = Memory.objects.get(pk=3)
memory  # 지정조건 즉시조회
```

```python
#bulk

memory_list =[]
for i in range(100):
    title = f'new title#{i}'
    content = f'new content#{i}'
    memory = Memory(title = title, message = content)
    # memory.save()  # 반복문에서 성능을 위해, 매번 save를 호출하지 않습니다
    
    memory_list.append(memory)
    
len(memory_list)
```

```python
Memory.objects.bulk_create(memory_list)
```

```python
Memory.objects.all().count()
```
